```python

df.reset_index(inplace = True ) # 기본 정수 Index
```

# Sort
- 특정 기준에 따라 순서대로 `나열`
- sort_value():특정 column 기준
```python
df.sort_values('col1', ascending = False, inplace = True )
```
  - ascending = False: 내림차순
  - ascending = True: 오름차순
  - inplace = True 를 하면 바뀐 상태로 df가 저장이 되고, 하지않으면 df를 다시 호출할 때는 처음 상태로 나타난다.

- sort_index(): index 기준
```python
df.sort_index(inplace= True)
```

````python
df_taxis.sort_values('distance', ascending = False, inplace = True)
#distance를 확인하면 큰거에서 작은 것으로 내려가는 것을 확인할 수 있음.
````

````python
df_taxis.head()
````

````python
#index가 뒤죽박죽 섞여 있기 때문에 다시 Index 로 정렬
df_taxis.sort_index(inplace = True)
````

````python
df_taxis.head()
````
