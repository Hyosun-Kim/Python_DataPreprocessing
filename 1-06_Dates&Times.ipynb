# to_period
- datetime으로 바꾼 것을 사용해서 주기를 변경하고, 원하는 형태로 데이터 포맷을 바꾸는 것을 할 것임
- 날짜/시간 데이터를 주어진 주기('M', 'W', 'Y' 등)로 변환
- dt는 날짜를 변경하는 method들이 들어있는 서랍이라고 생각하면됨
```python
 df['col'].dt.to_period('period').dt.strftime('%Y-%m-%d')
# to_period('period') : Daily(D), Weekly(W), Monthly(M), Yearly(Y) 등 주기로 나타냄
# dt.strftime('%Y-%m-%d') : 'YYYY-MM-DD' 형태로 표시함
 ```

````python
df.head()
df.info()
#pickup은 datetime으로 들어가 있고, dropoff는 object로 들어가있음
````

````python
df.pickup
#Weekly로 나타내기
df.pickup.dt.to_period('W').dt.strftime('%Y-%m-%d')
#pickup 으로 했을 때는 바뀌는데 아마 dropoff를 넣어서 돌린다면 안돌아감 왜냐하면 dropoff는 datetime 형태로 안바꿨기 때문.
#따라서 datetime으로 바꾸는 과정이 필요한거

#dt.to_timestamp() -> 시작 날짜를 보여줌
#dt.strftime('%Y-%m-%d') -> 끝나는 날짜를 보여줌
````

````python
#끝나는 날짜를 기준으로 보여주는 column 생성
df['pickup_w'] = df.pickup.dt.to_period('W').dt.strftime('%Y-%m-%d')
````

````python
df.info()
#pickup_w 를 확인해보면 잘 생성이 되었고, type을 보면 object로 생성되어있는 것을 알 수 있음
#pickup_w 는 일단 날짜 관련 전처리를 할 생각이 없기 때문에 그냥 놔두도록 하겠음
````

# timedelta
- 두 날짜 또는 시간 사이의 차이를 나타내는 데 사용
```python
df['duration'] = (df['end'] - df['start']).dt.total_seconds()
```

````python
df.dropoff - df.pickup
#timedelta라는 새로운 타입이 나옴, 편하게 보기 위해서 바꿔줄거임 ㄱㄷ
(df.dropoff - df.pickup).dt.total_seconds() #초로 나타내고 이것을 분 단위로 바꿔줄 거임(택시는 분 단위로 보는게 더 편해서)
(df.dropoff - df.pickup).dt.total_seconds()/60 #60으로 나눠주면 분단위가 나오겠죠?
````

start 열과 end 열이 있는 DataFrame df가 주어져요.
1. for 루프를 사용하여 DataFrame의 'start' 열과 'end' 열을 datetime 유형으로 변환하십시오.
2. start에서 end까지의 시간 차이를 시간 단위로 계산하십시오. 시간 이하 단위의 차이는 소수점으로 표시하십시오.
3. 'end' 열의 날짜들을 주 단위('W')로 변환하고, 이를 'yyyy-mm-dd' 형태의 문자열로 표현하십시오.



````python
#for 루프를 사용하여 DataFrame의 'start' 열과 'end' 열을 datetime 유형으로 변환하십시오.
# 데이터프레임 생성을 위한 딕셔너리
data = {
  'start': ['2021-01-01 10:00', '2021-01-02 11:30', '2021-01-03 09:45', '2021-01-04 15:00', '2021-01-05 18:20'],
  'end': ['2021-01-01 15:30', '2021-01-02 01:00', '2021-01-04 10:15', '2021-01-05 15:30', '2021-08-05 18:50']
}
# 데이터프레임 생성
df = pd.DataFrame(data)
df.info()
````

````python
pd.to_datetime(df.start)
#datetime으로 바꿈 but datetime으로 바꾼걸 저장해야되기 때문에 기존 데이터에 덮어씌우는 작업을 할 예정
df['start'] = pd.to_datetime(df.start)
df.info()
#여기서 우리가 처리해준거는 start만 datetime 형태로 처리해줬기 때문에 end도 for 루프를 통해서 처리해줄거임.
#columns 한 두개일 때는 직접 처리해줘도 되지만 엄청 많을 때는 for loop를 통해서 처리하는게 효율적임.
````

````python
col = df.columns
col
for i in col:
  df[i] = pd.to_datetime(df[i])
#i 라는 변수로 가져와서 모든 colums을 바꿔줄 수 있게 만들어주는거임.

#  df[pickup] = pd.to_datetime(df.pickup)
df.info()
````

````python
#start에서 end까지의 시간 차이를 시간 단위로 계산하십시오. 시간 이하 단위의 차이는 소수점으로 표시
(df['end'] - df['start']).dt.total_seconds()/3600
````

````python
#'end' 열의 날짜들을 주 단위('W')로 변환하고, 이를 'yyyy-mm-dd' 형태의 문자열로 표현
````

````python
df['end_weekly'] = df['end'].dt.to_period('W').dt.strftime('%Y-%m-%d')
````
