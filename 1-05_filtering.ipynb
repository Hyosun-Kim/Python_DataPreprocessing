# Filtering
- pd.DataFrame 에서 Filtering 은 조건에 부합(Boolean 값이 True)하는 row 만 남기는 과정
```python
df[조건]
# boolean-series:[True, False, ..., True]
```
  


````python
# 예시
import pandas as pd

# 예제 데이터프레임 생성
data = {
    'Name': ['철수', '영희', '흥민', '범근'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago']
}

df = pd.DataFrame(data)
````

````python
# df.Age >  30; age_over_30
con = df.Age > 30
df['age_over_30'] = con
````

````python
df
````

````python
# filter result
# df[con] df[df.Age > 30] 둘은 결국 같은 얘기
df[con]
````

````python
df[df.Age > 30]
````

1. 연습문제: 타이타닉 데이터셋 필터링
- 문제:아래는 타이타닉 호 승객 데이터에 대한 간략한 설명과 함께 필터링 작업을 수행하는 문제. Seaborn 라이브러리를 사용하여 타이타닉 데이터셋을 불러온 후, 다음 작업을 수행하시오.
  - class가 'First'이고, age가 30 이상인 승객들의 데이터를 필터링하세요.
  - survived가 1(생존)이거나 embark_town이 'Southampton'인 승객들의 데이터를 필터링하세요.
  - fare가 100 이상이고, sex가 'female'인 승객들의 데이터를 필터링하세요.



````python
##class가 'First'이고, age가 30 이상인 승객들의 데이터를 필터링하세요.
import seaborn as sns
df = sns.load_dataset('titanic')
````

````python
#df.class 라고 했더니 class를 다르게 인식해버림 파이썬에서 쓰는 기능으로 인식한 듯
df['class'].isin(['First'])
df['class'] == 'First'
````

````python
df.age > 30
````

````python
con = (df['class'] == 'First') & (df.age >= 30)
con
````

````python
df[con].age.value_counts()
````

````python
df[con].age.value_counts().sort_index() #가장 작은 값부터 차례로 보기 위함
````

````python
df[con]['class'].value_counts()
````

````python
#survived가 1(생존)이거나 embark_town이 'Southampton'인 승객들의 데이터를 필터링하세요.
````

````python
con2 = (df.survived ==1) | (df.embark_town == 'Southampton')
con2
````

````python
df[con2]
````

````python
df[con2].survived.value_counts()
````

````python
df[con2].embark_town.value_counts()
````

````python
df2 = df[con2]
````

````python
#검증
df2[df2.survived != 1].survived.value_counts()
df2[df2.survived != 1].embark_town.value_counts()
````

````python
#fare가 100 이상이고, sex가 'female'인 승객들의 데이터를 필터링하세요.
````

````python
con3 = (df.fare >= 100) & (df.sex == 'female')
con3
````

````python
df[con3]
````

````python
df[con3].fare.value_counts() #확실하게 모르겠으면 sort_index() 해서 작은 값을 맨위로 올려서 알아도됨
````

````python
df[con3].sex.value_counts()
````
